```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x_(N+1) = 4*X_(N)*(1 - X_(N))
```

```python
def funcao(x, lb):
  return 4*lb*x * (1 - x)
def d_funcao(x, lb):
  return 4*lb * (1 - x) - 4*lb*x

def recorrencia(x0, lb, max):
  iteracao = 0
  results = np.array([x0])
  while iteracao < max:
    x_n1 = funcao(x0, lb)
    x0 = x_n1
    results = np.append(results, x_n1)
    iteracao += 1
  return results

# Condições iniciais e valores de lambda:

x0  = 0.51
lb1 = 0.70
lb2 = 0.80
lb3 = 0.88
lb4 = 1.00
```

```python
max = 50
n = np.linspace(0, max + 1, max + 1)

arr1 = recorrencia(x0, lb1, max)
arr2 = recorrencia(x0, lb2, max)
arr3 = recorrencia(x0, lb3, max)
arr4 = recorrencia(x0, lb4, max)
```

```python
plt.figure(figsize=(15, 5))

plt.plot(n, arr1, label = r'$\lambda = 0.70$', color = "red",   marker = 's' )
plt.plot(n, arr2, label = r'$\lambda = 0.80$', color = "black", marker = 's')
plt.plot(n, arr3, label = r'$\lambda = 0.88$', color = "green", marker = 's')
plt.plot(n, arr4, label = r'$\lambda = 1.00$', color = "blue",  marker = 's')
plt.xlabel("Iteracoes (n)")
plt.ylabel("x_n")
plt.title("Comportamento da função discreta para diferentes valores de lambda")
plt.grid()
plt.legend()
plt.show()
```

Questão 2 -> Diagrama de teia


```python
# Define a função linear para o mapa de teia

def linear(x):
  return x

# Cria um vetor com as coordenadas para o mapa de teia

def pontos(arr):
  points = np.empty((0, 2))
  for i in range(len(arr) - 10):
    arr_1  = np.array([[arr[i], arr[i]]])
    arr_2  = np.array([[arr[i], arr[i+1]]])
    arr_12 = np.concatenate((arr_1, arr_2))
    points = np.concatenate((points, arr_12))
  return points

# Necessário para a criação dos gráficos

def coordenadas(vector_points, x0):
  points_x = np.array([x0])
  points_y = np.array([0])

  for i in range(len(vector_points)):
    points_x = np.append(points_x, vector_points[i][0])
    points_y = np.append(points_y, vector_points[i][1])
  return points_x, points_y




x                 = np.linspace(0, 1, 101)
y1, y2, y3, y4    = funcao(x, lb1),  funcao(x, lb2), funcao(x, lb3), funcao(x, lb4)    # Cria a função com os parâmetros lambda
y_l               = linear(x)                                                          # Cria a reta
p1, p2, p3, p4    = pontos(arr1), pontos(arr2), pontos(arr3), pontos(arr4)             # Chama a função para criar os pontos

coordp1_x, coordp1_y = coordenadas(p1, x0)
coordp2_x, coordp2_y = coordenadas(p2, x0)
coordp3_x, coordp3_y = coordenadas(p3, x0)
coordp4_x, coordp4_y = coordenadas(p4, x0)

# Plotagem dos dados

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Mapa de Teias para Diferentes Valores de $\lambda$', fontsize=14)

ax1.plot(x, y1, label=r'f(x) = $4x\lambda(1 - x)$', color='orange')
ax1.plot(x, y_l, label='f(x) = y', color='blue')
ax1.plot(coordp1_x, coordp1_y, 'o', color='red', linestyle='dashed', label=r'Mapa de teia para $\lambda$ = 0.70')
ax1.set_title(r"$\lambda$ = 0.70")
ax1.set_xlabel("y")
ax1.set_ylabel("x")
ax1.legend(loc = 'upper left', fontsize = 6)
ax1.grid(True)

ax2.plot(x, y2, label=r'f(x) = $4x\lambda(1 - x)$', color='orange')
ax2.plot(x, y_l, label='f(x) = y', color='blue')
ax2.plot(coordp2_x, coordp2_y, 'o', color='red', linestyle='dashed', label=r'Mapa de teia para $\lambda$ = 0.80')
ax2.set_title(r"$\lambda$ = 0.80")
ax2.set_xlabel("y")
ax2.set_ylabel("x")
ax2.legend(loc = 'upper left', fontsize = 6)
ax2.grid(True)

ax3.plot(x, y3, label=r'f(x) = $4x\lambda(1 - x)$', color='orange')
ax3.plot(x, y_l, label='f(x) = y', color='blue')
ax3.plot(coordp3_x, coordp3_y, 'o', color='red', linestyle='dashed', label=r'Mapa de teia para $\lambda$ = 0.88')
ax3.set_title(r"$\lambda$ = 0.88")
ax3.set_xlabel("y")
ax3.set_ylabel("x")
ax3.legend(loc = 'upper left', fontsize = 6)
ax3.grid(True)

ax4.plot(x, y4, label=r'f(x) = $4x\lambda(1 - x)$', color='orange')
ax4.plot(x, y_l, label='f(x) = y', color='blue')
ax4.plot(coordp4_x, coordp4_y, 'o', color='red', linestyle='dashed', label=r'Mapa de teia para $\lambda$ = 1.00')
ax4.set_title(r"$\lambda$ = 1.00")
ax4.set_xlabel("y")
ax4.set_ylabel("x")
ax4.legend(loc = 'upper left', fontsize = 6)
ax4.grid(True)

plt.tight_layout()
plt.show()
```

Questão 3 -> Mapa de recorrência

```python
def recorrencia_mapa(vetor_dados, erro):
  tam = len(vetor_dados)
  M   = np.zeros((tam, tam))
  for i in range(tam):
    for j in range(tam):
      if abs(vetor_dados[i] - vetor_dados[j]) < erro:
        M[i][j] = 1
      else:
        M[i][j] = 0
  return M

def getRval(matriz):
    x_vals, y_vals = [], []
    for i in range(len(matriz)):
        for j in range(len(matriz)):
            if matriz[i][j] == 1:
                x_vals.append(i)
                y_vals.append(j)
    return x_vals, y_vals

erro = 1e-4
# Gerando novamente os valores dos arrays de x_n. Isso é feito para observar melhor a recorrência

max_new  = 100
arr1_new = recorrencia(x0, lb1, max_new)
arr2_new = recorrencia(x0, lb2, max_new)
arr3_new = recorrencia(x0, lb3, max_new)
arr4_new = recorrencia(x0, lb4, max_new)

# Gerar a matriz de recorrência

M1 = recorrencia_mapa(arr1_new, erro)
M2 = recorrencia_mapa(arr2_new, erro)
M3 = recorrencia_mapa(arr3_new, erro)
M4 = recorrencia_mapa(arr4_new, erro)

coord1x, coord1y = getRval(M1)
coord2x, coord2y = getRval(M2)
coord3x, coord3y = getRval(M3)
coord4x, coord4y = getRval(M4)



# Plotando os pontos de recorrência

fig, ((ax11, ax22), (ax33, ax44)) = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Mapas de Recorrências para Diferentes Valores de $\lambda$', fontsize=14)

ax11.plot(coord1x, coord1y, 'o', ms = 1, color = 'red', label = r"Recorrência para $\lambda$ = 0.70")
ax11.set_title(r"$\lambda$ = 0.70")
ax11.set_xlabel("n")
ax11.set_ylabel("E")
ax11.legend(loc = 'lower right')
ax11.grid(True)

ax22.plot(coord2x, coord2y, 'o', ms = 1, color = 'black', label = r"Recorrência para $\lambda$ = 0.80")
ax22.set_title(r"$\lambda$ = 0.80")
ax22.set_xlabel("n")
ax22.set_ylabel("E")
ax22.legend(loc = 'lower right')
ax22.grid(True)

ax33.plot(coord3x, coord3y, 'o', ms = 1, color = 'green', label = r"Recorrência para $\lambda$ = 0.88")
ax33.set_title(r"$\lambda$ = 0.88")
ax33.set_xlabel("n")
ax33.set_ylabel("E")
ax33.legend(loc = 'lower right')
ax33.grid(True)

ax44.plot(coord4x, coord4y, 'o', ms = 1, color = 'blue', label = r"Recorrência para $\lambda$ = 1.00")
ax44.set_title(r"$\lambda$ = 1.00")
ax44.set_xlabel("n")
ax44.set_ylabel("E")
ax44.legend(loc = 'lower right')
ax44.grid(True)


plt.tight_layout()
plt.show()
```

Questão 4 -> Diagrama de bifurcação

```python
def coeficiente_lyapunov(n_max, x0, lb):
  sum = 0
  for i in range(n_max):
    sum += np.log(abs(d_funcao(x0, lb)))
    x0 = funcao(x0, lb)
  return sum/n_max



lambda_values     = np.linspace(0.80, 1, 1000)
max_int           = len(lambda_values)
x_n1              = np.zeros(max_int)
choose_points     = 1000

n_max = int(1e5)
lyapunov = coeficiente_lyapunov(n_max, x0, lambda_values)

fig, biax = plt.subplots()
fig.set_size_inches(16, 9)

for l in lambda_values:
  x_n1[0] = x0
  for j in range(max_int  - 1):
     x_n1[j+1] = funcao(x_n1[j], l)
  lambda_plot = [l]*choose_points
  xn1_plot = x_n1[max_int - choose_points:]
  biax.plot(lambda_plot, xn1_plot, "b.", markersize=0.03, color = 'red')

plt.plot(lambda_values, lyapunov, label = 'Coeficiente de Lyapunov', color = 'green')
plt.title(r"Mapa de bifurcação para $x_{0}$ = 0.51")
biax.set_xlabel(r"$\lambda$")
biax.set_ylabel(r"$x_{n+1}$")
plt.legend()
plt.show()
```
